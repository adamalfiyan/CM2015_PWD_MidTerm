## **3 Web Scraping the Data**

Firstly, before I web scrape the data, I will be defining all the libraries and tools that will be needed for the analysis:

```python
# Import necessary libraries and modules
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

### **3.1 Creating the web scraping functions**

As I will be web scraping data from multiple links on the website, I will be defining a function to fetch and parse the HTML content from a web page.

```python
def get_webpage_content(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    
    try:
        # Make the request
        response = requests.get(url, headers=headers, timeout=10)
        
        # Check if request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup
        else:
            print(f"Failed to retrieve the page. Status code: {response.status_code}")
            return None
            
    except requests.RequestException as e:
        print(f"Error during request: {str(e)}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {str(e)}")
        return None
```

#### **3.1.1 Extracting the Season Links**
In FBref, there is a table holding all the links for the different seasons in the Premier League. This function will be able to fetch me a number of links based on the range of seasons that I have specified.

```python
def get_season_links():
    standing_url = "https://fbref.com/en/comps/9/history/Premier-League-Seasons"
    
    try:
        soup = get_webpage_content(standing_url)
        
        if soup:
            standing_table = soup.select("table.stats_table")[0]
            links = standing_table.find_all("a")
            links = [l.get("href") for l in links]
            
            # Use a set to store unique links
            unique_links = set()
            for link in links:
                if '/comps/' in link:
                    season = link.split('/')[-2]
                    if '2014-2015' <= season <= '2023-2024':
                        full_link = f"https://fbref.com{link}"
                        unique_links.add((season, full_link))
            
            # Convert the set back to a sorted list
            season_links = sorted(unique_links, reverse=True)
            
            print(f"Found {len(season_links)} season links")
            for season, link in season_links:
                print(f"{season}: {link}")
                
            return season_links
            
        else:
            print("Failed to retrieve the webpage content")
            return None
            
    except Exception as e:
        print(f"Error occurred: {str(e)}")
        return None

season_links = get_season_links()
```

#### **3.1.2 Filtering the Links**

This function filters out the seasons that I do not want to include in my analysis, in this case the covid seasons 2019-2020 and 2020-2021.

```python
def filter_seasons(season_links, exclude_seasons=['2019-2020', '2020-2021']):

    if season_links is None:
        return None
        
    filtered_links = [(season, link) for season, link in season_links 
                     if season not in exclude_seasons]
    
    print(f"Filtered out {len(season_links) - len(filtered_links)} seasons")
    print(f"Remaining {len(filtered_links)} seasons (excluding {', '.join(exclude_seasons)})")
    for season, link in filtered_links:
        print(f"{season}: {link}")
        
    return filtered_links

# Then filter out the unwanted seasons
filtered_season_links = filter_seasons(season_links)
```

### **3.2 Extracting the Data from FBref**

Now that I have the links for the Premier League Seasons I want, I will be web scraping the data from them. Since there are multiple tables on each page, I will be creating a function to search for a specific table ID in each page and save the data. The web scraped data will be saved to a Pandas DataFrame first as it will be easier to manipulate and clean the data. It will be easier to inspect the data and handle missing values before finalising the data. 

I have noted that only some seasons have the advanced statistics such as "xG", "xGA", "xGD" and "xGD/90" which shows the expected goals/assists. These columns will be removed during the web scraping process as it is redundant and to ensure that all the data is aligned.

All the tables will be merged into one and saved into one CSV file as the data analysis will be comparing team performances and attendance from different seasons. This makes it easier to compare and spot patterns since I will be constantly referring to data from different seasons in the Premier League. 

This function scrapes the tables from the website and saves the data to a CSV file inside the current directory:

```python
def scrape_season_tables(filtered_season_links, table_ids):
    all_seasons_data = [] 
    headers = None  # To ensure consistent headers across seasons

    # Columns to ignore
    ignore_columns = {"xG", "xGA", "xGD", "xGD/90"}

    # Iterate through each season
    for season, url in filtered_season_links:
        print(f"\nProcessing season {season}...")

        if season not in table_ids:
            print(f"Skipping season {season}: No table ID found.")
            continue

        table_id = table_ids[season]
        soup = get_webpage_content(url) 
        if not soup:
            print(f"Skipping season {season}: Failed to retrieve webpage content.")
            continue

        table = soup.find('table', {'id': table_id})
        if not table:
            print(f"Skipping season {season}: Table with ID '{table_id}' not found.")
            continue

        # Extract headers
        header_row = table.find('tr')
        current_headers = [cell.text.strip() for cell in header_row.find_all(['th', 'td'])]
        filtered_headers = [header for header in current_headers if header not in ignore_columns]

        if headers is None:
            headers = ["Season"] + filtered_headers  # Initialise headers only once
            print(f"Headers initialized: {headers}")

        # Extract rows and filter columns
        rows = table.find_all('tr')[1:]  # Skip the header row
        for row in rows:
            cells = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
            filtered_row = [value for header, value in zip(current_headers, cells) if header not in ignore_columns]
            all_seasons_data.append([season] + filtered_row)

        print(f"Finished processing season {season}.")
        print("Waiting 3 seconds before the next request...")
        sleep(3)  # Add delay to avoid overwhelming the server

    # Create DataFrame
    premierleague_df = pd.DataFrame(all_seasons_data, columns=headers)
    print("\nData scraping complete. DataFrame created.")
    return premierleague_df


premierleague_df = scrape_season_tables(filtered_season_links, table_ids)
```

## **5. Exploratory Data Analysis**

### **5.1 Attendance Trends Over Seasons**

Firstly, we can take a look at how the average attendance has changed over the seasons using a line chart.

```python
# Calculate average attendance per season
season_attendance = premierleague_df.groupby("season")["attendance"].mean()

# Plot attendance trends over seasons
plt.figure(figsize=(10, 6))
season_attendance.plot(kind="line", marker="o", color="blue")
plt.title("Average Attendance Over Seasons")
plt.xlabel("Season")
plt.ylabel("Average Attendance")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

Here, we can see that the average attendance has increased over the years. This is likely due to the increase in popularity of the Premier League and the success of the teams that have played in the league. On top of that, the continuous developments of stadiums result in a higher seating capacity which can also affect the number of fans who attend the matches. It can also be seen that fan attendance hit an all time high during the 2021-2022 and 2022-2023 seasons. This is likely due to the COVID-19 pandemic which resulted in a sudden surge of fans wanting to watch their favourite teams play.

Now we can take a look at how the average attendance and win percentage are related for each team. We can use a bar chart to show the average attendance and a line chart to show the win percentage.

```python
# Group by team and calculate average attendance and win percentage
attendance_win_percentage = premierleague_df.groupby("team")[["attendance", "win_percentage"]].mean().sort_values("win_percentage", ascending=False)

# Plot average attendance and win percentage
plt.figure(figsize=(12, 6))
attendance_win_percentage["attendance"].plot(kind="bar", color="skyblue", label="Average Attendance")
plt.ylabel("Average Attendance")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.title("Average Attendance vs Win Percentage Per Team")

# Add a secondary axis for win percentage
plt.twinx()
attendance_win_percentage["win_percentage"].plot(kind="line", marker="o", color="orange", label="Win Percentage", linestyle="--")
plt.ylabel("Average Win Percentage (%)")
plt.legend(loc="upper left")

plt.show()
```

The graph displays a positive relationship between average attendance and win percentage, with teams that have higher attendance often achieving better win percentages. Teams such as Manchester United, Arsenal, and Liverpool, with attendance levels above 60,000, also maintain win percentages around 60–70%, reflecting the potential influence of strong fan support on team performance. On the other hand, teams with lower attendance, such as Sheffield United and Luton Town, often achieve win percentages below 20%, indicating that it could be challenging for smaller teams with limited fanbases. However, outliers such as West Ham and Newcastle United exhibit high attendance levels but maintain win percentages closer to mid-table teams, suggesting that attendance alone does not guarantee success.

### **5.3 Correlation between Fan Attendance and Team Performance**

#### **5.3.1 Scatter Plot**

I will be be starting my analysis with a scatter plot to visualise the relationship between fan attendance and points across different teams. Scatter plots are useful in showing the relationship between two variables and can help to quickly identify if they are positively or negatively correlated. They can also help to highlight trends and show the overall distribution of data points.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="attendance", y="points", data=premierleague_df, hue="season", palette="viridis", s=100, alpha=0.8)
plt.title("Attendance vs Points Across Teams")
plt.xlabel("Attendance")
plt.ylabel("Points")
plt.legend(title="Season", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid()
plt.show()
```
